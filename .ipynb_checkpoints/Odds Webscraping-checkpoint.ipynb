```python

# define a function to parse the information for each game
def parse_game_info(game_info, date):
    time = game_info[1]
    team1 = game_info[4]
    odds1 = game_info[10]
    team2 = game_info[9]
    odds2 = game_info[12]
    return {
        'date': date,
        'time': time,
        'team1': team1,
        'odds1': odds1,
        'team2': team2,
        'odds2': odds2
    }

# loop over each day's games and parse the information
games = []
for index, first_order_split in enumerate(string.split("B's")):
    if index > 0:
        second_order_split = first_order_split.split("\n")
        for i in range(int((len(second_order_split) - 5) / 14)):
            game_info = second_order_split[i*14:i*14+14]
            date = string.split("B's")[index-1].split("\n")[-4]
            game = parse_game_info(game_info, date)
            games.append(game)
LJL = pd.DataFrame(games)
LJL['league'] = "LJL"
LJL['date'] = pd.to_datetime(LJL['date'].str.slice(start=0, stop=11))

LJL.to_csv('data/LJL.csv')
```
